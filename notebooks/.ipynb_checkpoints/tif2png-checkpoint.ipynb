```python
import numpy as np
from glob import glob
from matplotlib import pyplot as plt
from skimage.io import imread, imsave
```

```python
def normalize(array):
    array_min, array_max = array.min(), array.max()
    return ((array - array_min)/(array_max - array_min))
```

```python
img = np.array(imread('../data/2018/haripur.465.tif'))
r = normalize(img[:,:,0])
g = normalize(img[:,:,1])
b = normalize(img[:,:,2])
rgb = np.dstack((r,g,b))
rgb = (rgb * 255).round().astype(np.uint8)
plt.imshow(rgb)
```

```python
img = np.array(imread('../data/2019/haripur.465.tif'))
r = normalize(img[:,:,0])
g = normalize(img[:,:,1])
b = normalize(img[:,:,2])
rgb = np.dstack((r,g,b))
rgb = (rgb * 255).round().astype(np.uint8)
plt.imshow(rgb)
```

```python
files = glob('../data'+'/**'+'/*.tif')
for f in files:
    y = f[8:12]
    pname = f[13:-4]
    img = np.array(imread(f))
    r = normalize(img[:,:,0])
    g = normalize(img[:,:,1])
    b = normalize(img[:,:,2])
    rgb = np.dstack((r,g,b))
    rgb = (rgb * 255).round().astype(np.uint8)
    imsave('../png/'+y+'/'+pname+'.png',rgb)
```
